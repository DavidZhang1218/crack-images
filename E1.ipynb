```python
import os
from PIL import Image

# æ ¹ç›®å½•
root_dir = r"C:\Users\admin\Desktop\Experiment01"

# éåŽ†æ‰€æœ‰å­æ–‡ä»¶å¤¹
for subfolder in os.listdir(root_dir):
    sub_path = os.path.join(root_dir, subfolder)
    if not os.path.isdir(sub_path):
        continue

    # åˆ›å»ºè¾“å‡ºæ–‡ä»¶å¤¹ï¼ˆè¦†ç›–åŽŸæ–‡ä»¶å¤¹å†…ï¼‰
    for filename in os.listdir(sub_path):
        file_path = os.path.join(sub_path, filename)
        if not os.path.isfile(file_path):
            continue

        # å°è¯•æ‰“å¼€å›¾åƒ
        try:
            with Image.open(file_path) as img:
                # è½¬æ¢ä¸ºRGBï¼ˆé¿å…RGBAæˆ–Pæ¨¡å¼å‡ºé”™ï¼‰
                img = img.convert("RGB")

                # æž„å»ºè¾“å‡ºè·¯å¾„ï¼ˆåŒåä½†æ‰©å±•åæ”¹ä¸º.jpgï¼‰
                base_name = os.path.splitext(filename)[0]
                output_path = os.path.join(sub_path, base_name + ".jpg")

                # ä¿å­˜ä¸ºJPG
                img.save(output_path, "JPEG", quality=95)
        except Exception as e:
            print(f"âŒ Failed to convert {file_path}: {e}")

print("âœ… All images have been converted to JPG format.")

```

```python
import os

# ç›®æ ‡æ–‡ä»¶å¤¹åˆ—è¡¨
folders = [
    r"C:\Users\admin\Desktop\Experiment01\Crack not on the road",
    r"C:\Users\admin\Desktop\Experiment01\No Crack",
    r"C:\Users\admin\Desktop\Experiment01\road_clean",
    r"C:\Users\admin\Desktop\Experiment01\road_crack"
]

# å…è®¸ä¿ç•™çš„æ‰©å±•åï¼ˆå°å†™ï¼‰
keep_exts = [".jpg", ".jpeg"]

for folder in folders:
    if not os.path.isdir(folder):
        print(f"âš ï¸ Skip non-folder path: {folder}")
        continue

    for filename in os.listdir(folder):
        file_path = os.path.join(folder, filename)
        if not os.path.isfile(file_path):
            continue

        ext = os.path.splitext(filename)[1].lower()
        if ext not in keep_exts:
            try:
                os.remove(file_path)
                print(f"ðŸ—‘ï¸ Deleted: {file_path}")
            except Exception as e:
                print(f"âŒ Failed to delete {file_path}: {e}")

print("âœ… Non-JPG images have been deleted.")
```

```python
import os

# å„ç±»åˆ«æ–‡ä»¶å¤¹è·¯å¾„
folders = {
    "road_without_crack": r"C:\Users\admin\Desktop\Experiment01\Road without Crack"
}

# å…è®¸çš„å›¾ç‰‡æ‰©å±•å
valid_exts = [".jpg", ".jpeg", ".png"]

for label, folder in folders.items():
    if not os.path.isdir(folder):
        print(f"âš ï¸ Skip non-folder path: {folder}")
        continue

    images = [f for f in os.listdir(folder) if os.path.splitext(f)[1].lower() in valid_exts]
    images.sort()  # æŽ’åºä»¥ä¿æŒä¸€è‡´

    for i, filename in enumerate(images, start=1):
        old_path = os.path.join(folder, filename)
        ext = os.path.splitext(filename)[1].lower()
        new_name = f"{label}_{i:04d}{ext}"
        new_path = os.path.join(folder, new_name)

        # é¿å…é‡åè¦†ç›–
        if old_path != new_path:
            os.rename(old_path, new_path)
            print(f"âœ… Renamed: {filename} â†’ {new_name}")

print("ðŸŽ¯ All images renamed successfully.")
```

```python
import os
import csv
import time
from openai import OpenAI

# =======================
# é…ç½®
# =======================
client = OpenAI(
    base_url='https://api-inference.modelscope.cn/v1',
    api_key='ms-38daaad6-6ada-49b5-86c7-db8a39246dd4',  # æ›¿æ¢ä¸ºä½ çš„ ModelScope Token
)

# GitHub raw URL åŸºç¡€è·¯å¾„
github_base = "https://raw.githubusercontent.com/DavidZhang1218/crack-images/main"

# å›¾ç‰‡ç±»åˆ«æ–‡ä»¶å¤¹ï¼ˆå’Œ GitHub ä¸Šä¿æŒä¸€è‡´ï¼‰
folders = [
    "Crack not on the road",
    "No Crack",
    "road_clean",
    "road_crack"
]

# è¾“å‡º CSV æ–‡ä»¶
output_csv = r"C:\Users\admin\Desktop\Experiment01\experiment_results.csv"

# è°ƒç”¨æ¨¡åž‹æ—¶çš„å»¶è¿Ÿï¼ˆç§’ï¼‰
CALL_DELAY = 1.5

# =======================
# Prompt æ¨¡æ¿
# =======================
def build_prompt():
    return """Given an image, please make two judgments and output ONLY a valid JSON object:
1. TaskA: Is the image mainly about a road or pavement scene? Answer "Yes" or "No".
2. TaskB: If TaskA = "Yes", is there any visible road crack on the pavement? Answer "Yes" or "No". If TaskA = "No", set TaskB to "N/A".
Also provide a short explanation (1â€“2 sentences).

Output strictly in JSON format, without any extra text.
Example:
{
  "taskA_is_road": "Yes",
  "taskB_has_road_crack": "No",
  "explanation": "The image shows a clean asphalt road surface with no visible cracks."
}"""

# =======================
# éåŽ†å›¾ç‰‡å¹¶è°ƒç”¨ Qwen3-VL
# =======================
results = []

# è¿™é‡Œå‡è®¾æœ¬åœ°æ–‡ä»¶å¤¹å’Œ GitHub æ–‡ä»¶å¤¹ä¸€è‡´ï¼Œç”¨äºŽè®°å½• ground truth
base_local = r"C:\Users\admin\Desktop\Experiment01"

for folder in folders:
    folder_path = os.path.join(base_local, folder)
    if not os.path.isdir(folder_path):
        print(f"âš ï¸ Skip non-folder path: {folder_path}")
        continue

    images = [f for f in os.listdir(folder_path) if f.lower().endswith((".jpg", ".jpeg", ".png"))]
    images.sort()

    for img_name in images:
        img_local_path = os.path.join(folder_path, img_name)
        # æ‹¼æŽ¥ GitHub raw URLï¼Œå¹¶å¤„ç†ç©ºæ ¼
        img_url = f"{github_base}/{folder}/{img_name}".replace(" ", "%20")

        print(f"ðŸ“Œ Processing: {img_url}")

        try:
            response = client.chat.completions.create(
                model='Qwen/Qwen3-VL-8B-Instruct',
                messages=[{
                    'role': 'user',
                    'content': [
                        {'type': 'text', 'text': build_prompt()},
                        {'type': 'image_url', 'image_url': {'url': img_url}}
                    ]
                }]
            )

            text = response.choices[0].message.content
            results.append({
                "image_path": img_local_path,
                "ground_truth_label": folder,
                "model_output": text
            })
            print(f"âœ… Done: {img_name}")

        except Exception as e:
            print(f"âŒ Failed: {img_name}, error: {e}")
            results.append({
                "image_path": img_local_path,
                "ground_truth_label": folder,
                "model_output": f"ERROR: {e}"
            })

        # å»¶è¿Ÿï¼Œé¿å…è§¦å‘é¢‘çŽ‡é™åˆ¶
        time.sleep(CALL_DELAY)

# =======================
# ä¿å­˜åˆ° CSV
# =======================
with open(output_csv, mode='w', newline='', encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=["image_path", "ground_truth_label", "model_output"])
    writer.writeheader()
    for row in results:
        writer.writerow(row)

print(f"ðŸŽ¯ Experiment finished. Results saved to: {output_csv}")
```
